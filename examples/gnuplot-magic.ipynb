Using Gnuplot in different language kernel
===========================================

This a Python 3 notebook, but we can switch on to use gnuplot
by way of gnuplot magic. Make sure you have looked at the
[gnuplot-kernel notebook](gnuplot-kernel.ipynb).

We can do the usual python stuff, but the key is the `%load_ext` magic
that we use to load `gnuplot_kernel` as an extension. This allows us
to use the gnuplot magic

```python
import matplotlib.pyplot as plt
import numpy as np

# inline plots for matplotlib
%matplotlib inline

# This loads the magics for gnuplot
%load_ext gnuplot_kernel
```

Plot using matplotlib

```python
np.random.seed(123)

N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N)) ** 2  # 0 to 15 point radii

plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```

Use the `%%gnuplot` cell magic for cells that contain gnuplot code.

```python
%%gnuplot
plot sin(x)
```

Use the `%gnuplot` line magic to change the inline plot options

```python
%gnuplot inline pngcairo size 600,400
```

```python
%%gnuplot
plot cos(x)
```
